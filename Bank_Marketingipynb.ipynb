```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

bank_train = bank_train.drop('ID', axis=1)
bank_test = bank_test.drop('ID', axis=1)
```

```python
cat_col = ['job', 'marital', 'education', 'default', 'balance', 'housing', 'loan', 'contact','month', 'poutcome']

le= LabelEncoder()
for p in cat_col:
  bank_train[p] = le.fit_transform(bank_train[p])
  bank_test[p] = le.fit_transform(bank_test[p])
```

```python
X_train = bank_train.drop('subscribed',axis=1)
y_train = bank_train['subscribed']
X_test = bank_test
```

```python
# Scaling Numerical features
sc = StandardScaler()
X_train[num_features] = sc.fit_transform(X_train[num_features])
X_test[num_features] = sc.fit_transform(X_test[num_features])
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)
```

```python
print(y_pred)
```

```python
#model evaluation

y_train_pred = rf.predict(X_train)
accuracy = accuracy_score(y_train, y_train_pred)
print("\nTraining Accuracy: ", accuracy)

print("\n Classification Report: \n", classification_report(y_train, y_train_pred))
```

```python
#cross validation to check overfitting

from sklearn.model_selection import cross_val_score

cv_score = cross_val_score(rf, X_train,y_train, cv=5, scoring = 'accuracy')
print("Cross Validation Score: ", cv_score)
print("Average Cross Validation Score: ", cv_score.mean())
```

Average Cross Validation Score indicates 90.55 model accuracy.


```python
from sklearn.metrics import roc_curve, roc_auc_score
y_train = le.fit_transform(y_train)
y_probs = rf.predict_proba(X_train)[:,1]

fpr, tpr, thresholds = roc_curve(y_train, y_probs)

auc = roc_auc_score(y_train, y_probs)

```

```python
plt.plot(fpr,tpr,label = 'ROC curve (area: %0.2f)' % auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel("True Positive Rate")
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```
