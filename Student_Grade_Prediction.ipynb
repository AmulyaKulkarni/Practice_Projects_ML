```python
null_col = ['CY-105', 'HS-105/12', 'MT-111', 'CS-106', 'EL-102', 'EE-119', 'ME-107',
       'CS-107', 'HS-205/20', 'MT-222', 'EE-222', 'MT-224', 'CS-210', 'CS-211',
       'CS-203', 'CS-214', 'EE-217', 'CS-212', 'CS-215', 'MT-331', 'EF-303',
       'HS-304', 'CS-301', 'CS-302', 'TC-383', 'MT-442', 'EL-332', 'CS-318',
       'CS-306', 'CS-312', 'CS-317', 'CS-403', 'CS-421', 'CS-406', 'CS-414',
       'CS-419', 'CS-423', 'CS-412']

for i in null_col:
  grades_df[i].fillna(grades_df[i].mode()[0],inplace=True)
```

```python

from sklearn.preprocessing import LabelEncoder

course_col = ['PH-121', 'HS-101', 'CY-105', 'HS-105/12', 'MT-111',
       'CS-105', 'CS-106', 'EL-102', 'EE-119', 'ME-107', 'CS-107', 'HS-205/20',
       'MT-222', 'EE-222', 'MT-224', 'CS-210', 'CS-211', 'CS-203', 'CS-214',
       'EE-217', 'CS-212', 'CS-215', 'MT-331', 'EF-303', 'HS-304', 'CS-301',
       'CS-302', 'TC-383', 'MT-442', 'EL-332', 'CS-318', 'CS-306', 'CS-312',
       'CS-317', 'CS-403', 'CS-421', 'CS-406', 'CS-414', 'CS-419', 'CS-423',
       'CS-412']

label_encoder = LabelEncoder()

for j in course_col:
  grades_df[j] = label_encoder.fit_transform(grades_df[j])
```

```python
features = grades_df.drop(['Seat No.', 'CGPA'],axis=1)
target = grades_df['CGPA']
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
X_train, X_test, y_train, y_test = train_test_split(features,target, test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2  = r2_score(y_test,y_pred)

print("Mean Squared Error: \n", mse)
print("Root Mean Squared Error: \n", rmse)
print("Mean Absolute Error: \n", mae)
print("R-squared value: \n", r2)
```

The evaluation metrics indicate good model performance.

```python
cross_val = cross_val_score(lr, features, target, cv=5)
print("Cross Validation Score: ",cross_val)
```

High Cross validation score indicates better model performance
