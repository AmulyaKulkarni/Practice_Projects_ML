```python
features = happy.drop('Happiness Score', axis=1)
target = happy['Happiness Score']
```

```python
print(features)
```

```python
print(target)
```

```python
features.columns
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

sc = StandardScaler()

cat_col = ['Country','Region']
num_col = ['Happiness Rank', 'Standard Error',
       'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)',
       'Freedom', 'Trust (Government Corruption)', 'Generosity',
       'Dystopia Residual']

preprocessor = ColumnTransformer(
    transformers=[
        ('n',sc, num_col),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_col)
    ]
)

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.3, random_state = 46)

X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

model_lr = LinearRegression()

model_lr.fit(X_train, y_train)

y_pred = model_lr.predict(X_test)
```

```python
#Visualising the model

plt.scatter(y_pred,y_pred, c='blue', label='Predicted')
plt.scatter(y_pred, y_test, c='green', label='Actual')

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Actual vs Predicted')

coef = np.polyfit(y_pred, y_test, 1)
ply = np.poly1d(coef)
x = np.linspace(min(y_pred), max(y_pred), 100)

plt.plot(x, ply(x),color='red', label='Best Fit Line')
plt.legend()
plt.show()
```

Evaluation Metrics for the Model

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse_lr = mean_squared_error(y_test, y_pred)
rmse_lr = np.sqrt(mean_squared_error(y_test,y_pred))
mae_lr = mean_absolute_error(y_test,y_pred)
r2_lr = r2_score(y_test,y_pred)

print("Mean Squared Error: ",mse_lr)
print("Root Mean Squared Error: ",rmse_lr)
print("Mean Absolute Error: ",mae_lr)
print("R2 Score: ",r2_lr)
```

From the scores above, the MSE, RMSE, MAE are quite low and the R2 Score is high. This suggests a better model performance. But we will be checking the model for overfitting

## Checking the model for overfitting

```python
# Accessing the model for overfitting

from sklearn.model_selection import cross_val_score, learning_curve

#Cross Validation Score
cv_score = cross_val_score(model_lr, X_train, y_train, cv=5)
print("Cross Validation Score: ", cv_score)
print("Mean Cross Validation Score: ", cv_score.mean())
```

The Cross Validation Score indicates high level of accuracy

```python
from sklearn.model_selection import learning_curve

train_size, train_score, val_score = learning_curve(model_lr, X_train, y_train, cv=5)

train_mean = np.mean(train_score, axis=1)
train_std = np.std(train_score,axis=1)
val_mean = np.mean(val_score,axis=1)
val_std = np.std(val_score,axis=1)

plt.figure(figsize=(10,8))
plt.plot(train_size,train_mean, label='Training Score')
plt.plot(train_size,val_mean,label='Validation Score')
plt.fill_between(train_size, train_mean - train_std, train_mean + train_std, alpha = 0.2)
plt.fill_between(train_size, val_mean - val_std, val_mean + val_std, alpha = 0.2)
plt.xlabel('Training Set Size')
plt.ylabel('Score')
plt.title('Learning Curves')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import Ridge

ridge_m = Ridge(alpha=0.1)

ridge_m.fit(X_train,y_train)

y_pred_ridge = ridge_m.predict(X_test)
mse_ridge = mean_squared_error(y_test,y_pred_ridge)
rmse_ridge = np.sqrt(mse_ridge)
mae_ridge = mean_absolute_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

print("Mean Squared Error: ",mse_ridge)
print("Root Mean Squared Error: ",rmse_ridge)
print("Mean Absolute Error: ",mae_ridge)
print("R2 Score: ",r2_ridge)
```

The above metrics indicate high model accuracy
