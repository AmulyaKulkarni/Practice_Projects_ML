Data Preprocessing and Feature Engineering

```python
from sklearn.preprocessing import LabelEncoder
e = LabelEncoder()
titanic['Sex'] = e.fit_transform(titanic['Sex'])
titanic['Embarked'] = e.fit_transform(titanic['Embarked'])
```

```python
titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch'] + 1
titanic.head()
```

```python
features = ['Pclass', 'Sex', 'Age', 'Embarked', 'FamilySize']
X = titanic[features]
y = titanic['Survived']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=45)
```

```python
from sklearn.linear_model import LogisticRegression
log_r = LogisticRegression()
log_r.fit(X_train, y_train)
y_pred = log_r.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, f1_score

accuracy = accuracy_score(y_test,y_pred)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)
con_mat = confusion_matrix(y_test,y_pred)

print("Accuracy: ", accuracy)
print("Precision: ", precision)
print("Recall: ", recall)
print("F1 Score: ", f1)
print("Confusion Matrix: \n",con_mat)
```

```python
sb.heatmap(con_mat, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.xticks([0.5,1.5],['Not Survived', 'Survived'])
plt.yticks([0.5,1.5],['Not Survived', 'Survived'])

plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

y_prob = log_r.predict_proba(X_test)[:,1]
fpr,tpr, threhsolds = roc_curve(y_test, y_prob)
auc = roc_auc_score(y_test,y_prob)

plt.plot(fpr, tpr, label='ROC curve (AUC = {:.2f})'.format(auc))
plt.plot([0,1],[0,1], '--', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')

plt.show()
```
